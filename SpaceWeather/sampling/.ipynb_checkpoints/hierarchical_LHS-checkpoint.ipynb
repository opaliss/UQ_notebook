## Experiment Design: Hierarchical Latin Hypercube Sampler

```python
import numpy as np 
import scipy.stats.qmc as qmc
from skopt.sampler import Lhs
from skopt.sampler import Grid
from scipy.spatial.distance import pdist
from sklearn.utils import check_random_state
from scipy import spatial
from skopt.sampler.base import InitialPointGenerator
from skopt.space import Space, Categorical
from HLHS import LHS_extendable
import matplotlib.pyplot as plt

import matplotlib
font = {'family' : 'serif',
        'size'   : 18}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 
```

```python
# define 1D uniform [0, 1] sampler to be LHS
sampler = qmc.LatinHypercube(d=1, centered=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(sampler.random(n=1), 1, c="k")

Nold = 1

for r in range(2, 5):
    # old samples
    samples_old = sampler.random(n=Nold)
    # update Nold
    Nold = 3*Nold
    # sample again
    samples_new = sampler.random(n=Nold)
    for sample in samples_new:
        if sample in samples_old:
            _ = ax.scatter(sample, r, c="r")
        else:
            _ = ax.scatter(sample, r, c="k")
            
_ = ax.scatter(None, None, c="k", label="New Samples")
_ = ax.scatter(None, None, c="r", label="Old Samples")
_ = plt.gca().invert_yaxis()
_ = plt.legend()
_ = ax.set_xlim(0, 1)

_ = ax2 = ax.twinx()
_ = ax2.set_ylabel("Number of Added Samples")
_ = ax2.set_ylim(ax.get_ylim())
_ = ax2.set_yticks([4, 3, 2, 1])
_ = ax2.set_yticklabels(["18", "6", "2", "0"])

_ = ax.set_xticks(np.linspace(0, 1, 10))
_ = ax.set_yticks([1, 2, 3, 4])
_ = ax.set_yticklabels(["1", "3", "9", "27"])
_ = ax.grid(axis="y")
_ = ax.grid(axis="x")
_ = ax.set_ylabel("Total Number of Samples")
_ = ax.set_xlabel("LHS Samples")
_ = ax.set_title("1D Hierarchical Latin Hypercube Sample")
plt.savefig("figs/hlhs_2d.png", dpi=500)
```

```python
# # define 2D uniform [0, 1] sampler to be LHS
sampler = qmc.LatinHypercube(d=2, centered=True)
```

```python
def check_if_in_set(x, y):
    for sample in range(np.shape(y)[0]):
        #print(np.linalg.norm(np.abs(x - y[sample, :])))
        if np.linalg.norm(np.abs(x - y[sample, :])) < 1e-9 :
            return True
    return False

def latin_hypercube_2d_uniform(n):
    lower_limits = np.arange(0, n)/n
    upper_limits = np.arange(1, n+1)/n
    pointsx = (lower_limits + upper_limits)/2
    points = np.zeros((n, 2))
    points[:, 0] = (lower_limits + upper_limits)/2
    points[:, 1] = (lower_limits + upper_limits)/2

    #np.random.shuffle(points[:, 1])
    return points
```

```python
fig, ax = plt.subplots(nrows=4, figsize=(5, 15))

sample = sampler.random(n=1)[0]
ax[0].scatter(sample[0],sample[1], c="k")

ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 1)

Nold = 1

for r in range(2, 5):
    # old samples
    samples_old = latin_hypercube_2d_uniform(n=Nold)
    # update Nold
    Nold = 3*Nold
    # sample again
    samples_new = latin_hypercube_2d_uniform(n=Nold)
    
    for sample in samples_new:
        if check_if_in_set(sample, samples_old):
            _ = ax[r-1].scatter(sample[0],sample[1], c="r")
            if r < 4:
                _ = ax[r].scatter(sample[0],sample[1], c="b")
        else:
            _ = ax[r-1].scatter(sample[0],sample[1], c="k")
            if r < 4:
                _ = ax[r].scatter(sample[0],sample[1], c="b")

    _ = ax[r-1].set_xlim(0, 1)
    _ = ax[r-1].set_ylim(0, 1)
    _ = ax[r-1].set_xticks(np.linspace(0, 1, Nold+1))
    _ = ax[r-1].set_yticks(np.linspace(0, 1, Nold+1))
    _ = ax[r-1].grid(axis="y")
    _ = ax[r-1].grid(axis="x")
    _ = ax[r-1].set_xticklabels([])
    _ = ax[r-1].set_yticklabels([])
            
_ = ax[0].scatter(None, None, c="k", label="New Samples")
_ = ax[0].scatter(None, None, c="r", label="Old Samples")

_ = ax[0].legend(loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.3))


plt.savefig("figs/hlhs_2d.png", dpi=500)
```

```python
space = Space([(0., 1.), (0., 1.)])
lhs = LHS_extendable(criterion="maximin", iterations=10000)
```

```python
fig, ax = plt.subplots(nrows=4, figsize=(5, 15))

sample = lhs.generate(space.dimensions, 1)[0]
ax[0].scatter(sample[0],sample[1], c="k")

ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 1)
_ = ax[0].set_xticklabels([])
_ = ax[0].set_yticklabels([])
Nold = 1
 
samples_old = lhs.generate(space.dimensions, 1)

for r in range(2, 5):
    # update Nold
    Nold = 3*Nold
    # sample again
    samples_new = lhs.generate(space.dimensions, Nold, existing_samples = np.array(samples_old))
    
    for sample in samples_new:
        if check_if_in_set(sample, np.array(samples_old)):
            _ = ax[r-1].scatter(sample[0],sample[1], c="r")
            if r < 4:
                _ = ax[r].scatter(sample[0],sample[1], c="b")
        else:
            _ = ax[r-1].scatter(sample[0],sample[1], c="k")
            if r < 4:
                _ = ax[r].scatter(sample[0],sample[1], c="b")
    samples_old = samples_new

    _ = ax[r-1].set_xlim(0, 1)
    _ = ax[r-1].set_ylim(0, 1)
    _ = ax[r-1].set_xticks(np.linspace(0, 1, Nold+1))
    _ = ax[r-1].set_yticks(np.linspace(0, 1, Nold+1))
    _ = ax[r-1].grid(axis="y")
    _ = ax[r-1].grid(axis="x")
    _ = ax[r-1].set_xticklabels([])
    _ = ax[r-1].set_yticklabels([])
    _ = ax[r-1].set_title("Number of Samples = "+ str(Nold))
            
_ = ax[0].scatter(None, None, c="k", label="New Samples")
_ = ax[0].scatter(None, None, c="r", label="Old Samples")

_ = ax[0].legend(loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.3))


plt.savefig("figs/hlhs_2d.png", dpi=500)
```
